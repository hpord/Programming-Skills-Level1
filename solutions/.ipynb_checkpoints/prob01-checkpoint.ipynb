```python
class Jugador:
    def __init__(self, name, dorsal, goals, speed, assists, passing_acc, defensive):
        self._name = name
        self._dorsal = dorsal
        self._goals = goals
        self._speed = speed
        self._assists = assists
        self._passing_acc = passing_acc
        self._defensive = defensive
        
    @property
    def name(self):
        return self._name
    @property
    def dorsal(self):
        return self._dorsal
    @property
    def goals(self):
        return self._goals
    @property
    def speed(self):
        return self._speed
    @property
    def assists(self):
        return self._assists
    @property
    def passing_acc(self):
        return self._passing_acc
    @property
    def defensive(self):
        return self._defensive
```

```python
class Analizador:
    def __init__(self, jugadores):
        self._jugadores = jugadores
        self._goleador = None
        self._mas_rapido = None
        self._mas_asistidor = None
        self._mas_preciso = None
        self._mas_defensivo = None
        self.buscar_goleador()
        self.buscar_mas_rapido()
        self.buscar_mas_asistidor()
        self.buscar_mas_preciso()
        self.buscar_mas_defensivo()
        self._opciones = [
            None,
            self.revisión_jugador,
            self.comparar_jugadores,
            self.mostrar_mas_rapido,
            self.mostrar_goleador,
            self.mostrar_mas_asistidor,
            self.mostrar_mas_preciso,
            self.mostrar_mas_defensivo,
            self.salir
        ]
        
    def analizar(self):
        while True:
            print("\nBienvenido al analizador de jugadores del Manchester United \n")
            print("Seleccione la acción que desea realizar:")
           # print("(Digite el número correspondiente)")
            print("----------------------")
            print("1. Revisar jugador")
            print("2. Comparar dos jugadores")
            print("3. Identificar al más rápido")
            print("4. Identificar al máximo goledor")
            print("5. Identificar al jugador con mayores asistencias")
            print("6. Identificar al jugador con mayor precisión de pases")
            print("7. Identificar al jugador con mayores intervenciones defensivas")
            print("8. Salir\n")
            opcion = 0
            while opcion < 1 or opcion > 8:
                try:
                    opcion = int(input("Digite el número correspondiente a la acción: "))
                    print()
                except:
                    print("Entrada no válida")
                    opcion = 0
                if opcion < 1 or opcion > 8:
                    print("Valores fuera del rango")
                    print("Ingrese valores correctos")
            self._opciones[opcion]()
            if opcion != 8:
                eleccion = "z"
                while eleccion.lower() not in ["si", "s", "no", "n"]:
                    eleccion = input("¿Desea volver al menú principal?(Si/No): ")
                    print()
                    if eleccion.lower() not in ["si", "s", "no", "n"]:
                        print("Entrada no válida")
                        print("Ingrese un valor correcto")
                if eleccion.lower() in ["no", "n"]:
                    opcion = 8
                    self.salir()
            if opcion == 8: break
            
    
    def buscar_goleador(self):
        goles = -1
        for jugador in jugadores.values():
            if jugador.goals > goles:
                self._goleador = [jugador]
                goles = jugador.goals
            elif jugador.goals == goles:
                self._goleador.append(jugador)
        
    def buscar_mas_rapido(self):
        rapidez = -1
        for jugador in jugadores.values():
            if jugador.speed > rapidez:
                self._mas_rapido = [jugador]
                rapidez = jugador.speed
            elif jugador.speed == rapidez:
                self._mas_rapido.append(jugador)
        
    def buscar_mas_asistidor(self):
        ptos_asist = -1
        for jugador in jugadores.values():
            if jugador.assists > ptos_asist:
                self._mas_asistidor = [jugador]
                ptos_asist = jugador.assists
            elif jugador.assists == ptos_asist:
                self._mas_asistidor.append(jugador)
        
    def buscar_mas_preciso(self):
        ptos_prec = -1
        for jugador in jugadores.values():
            if jugador.passing_acc > ptos_prec:
                self._mas_preciso = [jugador]
                ptos_prec = jugador.passing_acc
            elif jugador.passing_acc == ptos_prec:
                self._mas_preciso.append(jugador)
        
    def buscar_mas_defensivo(self):
        ptos_def = -1
        for jugador in jugadores.values():
            if jugador.defensive > ptos_def:
                self._mas_defensivo = [jugador]
                ptos_def = jugador.defensive
            elif jugador.defensive == ptos_def:
                self._mas_defensivo.append(jugador)
    
    def comparar_jugadores(self):
        dorsal1, dorsal2 = 0, 0
        print("Comparación de jugadores")
        while dorsal1 not in jugadores or dorsal2 not in jugadores:
            dorsal1 = int(input("Ingrese número del primer jugador: "))
            dorsal2 = int(input("Ingrese número del segundo jugador: "))
            print()
            if dorsal1 not in jugadores or dorsal2 not in jugadores:
                print("Numero(s) de camiseta(s) no encontrado(s)")
                print("Vuelva a ingresar números de camiseta válidos")
        long1, long2 = len(jugadores[dorsal1].name) + 3, len(jugadores[dorsal2].name) + 3
        cadena0 = f"Nombre    "
        cadena1 = f"{: >long1}".format(jugadores[dorsal1].name)
        cadena2 = f"{: >long2}".format(jugadores[dorsal2].name)
        print(cadena0, cadena1, cadena2)
        cadena0 = f"Goles".format(10)
        cadena1 = f"{jugadores[dorsal1].goals}".format(long1)
        cadena2 = f"{jugadores[dorsal2].goals}".format(long2)
        print(cadena0, cadena1, cadena2)
        cadena0 = f"Rapidez".format(10)
        cadena1 = f"{jugadores[dorsal1].speed}".format(long1)
        cadena2 = f"{jugadores[dorsal2].speed}".format(long2)
        print(cadena0, cadena1, cadena2)
        cadena0 = f"Asistencia".format(10)
        cadena1 = f"{jugadores[dorsal1].assists}".format(long1)
        cadena2 = f"{jugadores[dorsal2].assists}".format(long2)
        print(cadena0, cadena1, cadena2)
        cadena0 = f"Precisión".format(10)
        cadena1 = f"{jugadores[dorsal1].passing_acc}".format(long1)
        cadena2 = f"{jugadores[dorsal2].passing_acc}".format(long2)
        print(cadena0, cadena1, cadena2)
        cadena0 = f"Defensa".format(10)
        cadena1 = f"{jugadores[dorsal1].defensive}".format(long1)
        cadena2 = f"{jugadores[dorsal2].defensive}".format(long2)
        print(cadena0, cadena1, cadena2)
        print()
    
    def revisión_jugador(self):
        print("Revisión de jugador")
        dorsal = 0
        while dorsal not in jugadores:
            dorsal = int(input("Ingrese número del jugador: "))
            print()
            if dorsal not in jugadores:
                print("Numero de camiseta no encontrado")
                print("Vuelva a ingresar un número de camiseta válido")
        print(f"El jugador {jugadores[dorsal].name} con la camiseta {dorsal} tiene las siguientes características: ")
        print(f"Lleva {jugadores[dorsal].goals} gol(es) anotado(s)")
        print(f"Tiene {jugadores[dorsal].speed} en velocidad")
        print(f"Tiene {jugadores[dorsal].assists} en asistencias")
        print(f"Tiene {jugadores[dorsal].speed} en velocidad")
        print(f"Tiene {jugadores[dorsal].passing_acc} en precisión de pases")
        print(f"Tiene {jugadores[dorsal].defensive} en intervenciones defensivas")
        print()
        
    def mostrar_mas_rapido(self):
        if len(self._mas_rapido) == 1:
            print(f"El jugador más rápido es {self._mas_rapido[0].name} con {self._mas_rapido[0].speed} puntos de rapidez")
        else:
            print(f"Los jugadores más rápidos son: ")
            for rapido in self._mas_rapido:
                print(f"- {rapido.name}")
            print(f"Ellos tienen {self._mas_rapido[0].speed} puntos de rapidez")
        print()
        
    def mostrar_goleador(self):
        if len(self._goleador) == 1:
            print(f"El jugador más goleador es {self._goleador[0].name} con {self._goleador[0].goals} goles")
        else:
            print(f"Los jugadores más goleadores son: ")
            for goleador in self._goleador:
                print(f"- {goleador.name}")
            print(f"Ellos llevan {self._goleador[0].goals} goles")
        print()
        
    def mostrar_mas_asistidor(self):
        if len(self._mas_asistidor) == 1:
            print(f"El jugador más asistidor es {self._mas_asistidor[0].name} con {self._mas_asistidor[0].assists} puntos de asistencia")
        else:
            print(f"Los jugadores más asistidores son: ")
            for asistidor in self._mas_asistidor:
                print(f"- {asistidor.name}")
            print(f"Ellos tienen {self._mas_asistidor[0].assists} puntos de asistencia")
        print()
    
    def mostrar_mas_preciso(self):
        if len(self._mas_preciso) == 1:
            print(f"El jugador con mejor precisión de pases es {self._mas_preciso[0].name} con {self._mas_preciso[0].passing_acc} puntos de precisión de pases")
        else:
            print(f"Los jugadores con mejor precisión de pases son: ")
            for preciso in self._mas_preciso:
                print(f"- {preciso.name}")
            print(f"Ellos tienen {self._mas_preciso[0].passing_acc} puntos de precisión de pases")
        print()
    
    def mostrar_mas_defensivo(self):
        if len(self._mas_defensivo) == 1:
            print(f"El jugador con mayores intervenciones defensivas es {self._mas_defensivo[0].name} con {self._mas_defensivo[0].defensive} puntos de intervenciones defensivas")
        else:
            print(f"Los jugadores con mayores intervenciones defensivas son: ")
            for defensor in self._mas_defensivo:
                print(f"- {defensor.name}")
            print(f"Ellos tienen {self._mas_defensivo[0].defensive} puntos de intervenciones defensivas")
        print()
    def salir(self):
        print("Gracias por su confianza")
        print("Hasta pronto\n") 
    
```

```python
bruno_fernandes = Jugador("Bruno Fernandes", 8, 5, 6, 9, 10, 3)
rasmus_hojlund = Jugador("Rasmus Hojlund", 11, 12, 8, 2, 6, 2)
harry_maguire = Jugador("Harry Maguire", 5, 1, 5, 1, 7, 9)
alejandro_garnacho = Jugador("Alejandro Garnacho", 17, 8, 7, 8, 6, 0)
mason_mount = Jugador("Mason Mount", 7, 2, 6, 4, 8, 1)

jugadores = {
    bruno_fernandes.dorsal: bruno_fernandes,
    rasmus_hojlund.dorsal: rasmus_hojlund,
    harry_maguire.dorsal: harry_maguire,
    alejandro_garnacho.dorsal: alejandro_garnacho,
    mason_mount.dorsal: mason_mount
}

analizador = Analizador(jugadores)
```

```python
def main():
    analizador.analizar()
```

```python
if __name__ == "__main__":
    main()
```
